# *Numpy* 学习

## 1.导包

```python
import numpy as np
```

## 2.查看属性

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print('数组形状:', arr.shape)
print('元素个数:', arr.size)
print('数组维度：', arr.ndim)
print('元素数据类型:', arr.dtype)
print('转置:', arr.T)
```

## 3.*ndarry* 的创建

```python
list1 = [1,2,3],[4,5,6]
arr = np.array([list1],dtype = np.float64) #强制类型转换
print(arr)
```

### copy

```python
# copy
list2 = np.copy(list1)
print(list2)
arr2 = np.copy(arr)
print(arr2)
```

### 预定义形状

```python
# 预定义形状，初始化数组
# 全0
arr3 = np.zeros_like(arr2)
print(arr3)
arr4 = np.zeros((2,4))
print(arr4)
# 全1
arr5 = np.ones_like(arr2, dtype = np.int64)
print(arr5)
# 初始化为null
arr6 = np.empty_like(arr2)
print(arr6)
```

### 数列

```python
# 等差数列
arr7 = np.arange(1, 6,2) #开始，结束，步长
print(arr7)
```

```python
# 等间隔数列
arr8 = np.linspace(0,10,5,dtype = np.int64)
print(arr8)
```
