# *Numpy* 学习

## 1.导包

```python
import numpy as np
```

## 2.查看属性

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print('数组形状:', arr.shape)
print('元素个数:', arr.size)
print('数组维度：', arr.ndim)
print('元素数据类型:', arr.dtype)
print('转置:', arr.T)
```

## 3.*ndarry* 的创建

```python
list1 = [1,2,3],[4,5,6]
arr = np.array(list1,dtype = np.float64) #强制类型转换
print(arr)
print(arr.ndim)
```

### copy

```python
# copy
list2 = np.copy(list1)
print(list2)
arr2 = np.copy(arr)
print(arr2)
```

### 预定义形状

```python
# 预定义形状，初始化数组
# 全0
arr3 = np.zeros_like(arr2)
print(arr3)
arr4 = np.zeros((2,4))
print(arr4)
# 全1
arr5 = np.ones_like(arr2, dtype = np.int64)
print(arr5)
# 初始化为null
arr6 = np.empty_like(arr2)
print(arr6)
```

### 数列

```python
# 等差数列
arr7 = np.arange(1, 6,2) #开始，结束，步长
print(arr7)
```

```python
# 等间隔数列
arr8 = np.linspace(0,10,5,dtype = np.int64)
print(arr8)
```

### 特殊矩阵

```python
# 单位矩阵
arr = np.eye(3,4,dtype=np.int64)
print(arr)
```

```python
# 对角矩阵
list1 = [1,2,3]
arr = np.diag(list1)
print(arr)
```

```python
# 随机数组生成
## 随机浮点数（0-1）
arr = np.random.rand(2,3)
print(arr)
## 随机浮点数（指定区间）
arr2 = np.random.uniform(3,4,(2,3))
print(arr2)
## 随机整数
arr3 = np.random.randint(1,5,(2,3))
print(arr3)
## 随机数列（正态分布）
arr4 = np.random.randn(2,3)
print(arr4)
```

```python
## 设置随机种子
## 随机种子是一套固定的数学公式
## 只要种子数字一样
## 算出来的结果完全相同
np.random.seed(20)
arr5 = np.random.randint(1,5,(2,3))
print(arr5)
print(arr5)
```

## 4.*Ndarry* 的数据类型

```python
## 布尔类型
list1 = [1,0,0,1]
arr = np.array(list1,dtype = np.bool)
print(arr)

## 整数类型
list2 = [1,127,0,1]
arr = np.array(list2,dtype = np.int8) #8位有符号的整数范围为-128至127
print(arr)
list2 = [1,250,0,255]
arr = np.array(list2,dtype = np.uint8) #8位无符号的整数范围为0至255
print(arr)
```
