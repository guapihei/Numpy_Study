# Series

## 1.Series的创建

```python
# series的创建
import numpy as np
import pandas as pd
s = pd.Series([10,2,3,4,5]) #带了索引
# 自定义索引
s = pd.Series([10,2,3,4,5],index=['a','b','c','d','e'])
print(s)
#s = pd.Series([10,2,3,4,5],name='name')
#print(s)
s1 = pd.Series(s,index=["a","c"])
print(s1)
```

## 2.Series的属性

```python
# Series的相关属性
print(s.index) # 索引
print(s.values) # 值
print(s.shape, s.ndim, s.size) # 形状、维度、大小
s.name='年份'
print(s.dtype, s.name) # 数据类型、标签
print(s.loc['a']) # 通过显式索引查询
print(s.iloc[1]) # 通过隐式索引（位置）查询
print(s.loc['a':'d']) # 索引支持切片（显式索引、隐式索引都支持切片）
print(s.at['a']) # 显式获取，at不支持切片
print(s.iat[1]) # 隐式获取
```

```python
# 常见函数
import numpy as np
s = pd.Series([10, 2, 3, np.nan, 4, None, 2, 4], index=['a','b','c','d','e','f','g','h'], name='金额')
print(s)
s.head(3) # 获取前3行
s.tail(2) # 获取最后2行
s.describe() # 获取统计信息
print(s.count()) # 获取元素个数（会忽略缺失值）
print(s.keys()) # 通过keys也可以获取索引，和s.index属性结果相同
s.isna() # 检查Series中的每个元素是否为缺失值
s.isin([4,5,6]) # 检查Series的元素是否在指定的列表中
print(s.quantile(0.25)) # pd中的分位数的方法是quantile，传入的是小数，和numpy的不同
print(s.drop_duplicates()) # 去除重复项，返回的还是Series
print(s.unique()) # 去除重复项，返回一个array
print(s.sort_index()) # 按索引排序
print(s.sort_values()) # 按值排序
```

```python
# 创建一个包含10名学生数学成绩的Series，成绩范围在50-100。
## 计算平均分、最高分、最低分
## 找出高于平均分的学生人数
np.random.seed(20)

chengji = pd.Series(np.random.randint(50,101,10))
print(chengji)
pjf = chengji.mean()
chengji[chengji >pjf].count()
```

```python
# 某商店每小时销售额Series：
## 按天重采样计算每日总销售额
## 计算每天营业时间(8:00-22:00)和非营业时间的销售额比例
## 找出销售额最高的3个小时
np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(10, 100, 24),
                         index=pd.date_range(start='2025-01-01', periods=24,freq='h'),
                         name='销售额')

print('每日总销售额: \n', hourly_sales.resample('D').sum())

# 营业时间的销售额
a = hourly_sales.between_time('8:00', '22:00') # 直接使用 between_time 获取时间窗口
a1 = hourly_sales[(hourly_sales.index.hour >= 8) & (hourly_sales.index.hour <= 22)] # 第2种方式：通过布尔索引，截取8点到22点。hourly_sales.index 是个时间，可以获取hour、year等属性
# 非营业时间的销售额
b = hourly_sales.sum() - a.sum() # 直接用总金额减去营业时间的金额
b1 = hourly_sales.drop(a.index) # 第2种方式：使用drop函数，从总Series中减去营业时间的索引
b2 = hourly_sales[(hourly_sales.index.hour < 8) | (hourly_sales.index.hour > 22)] # 第3种方式：通过布尔索引方式
print('每日营业时间和非营业时间的销售额占比: ', a.sum() / b)

print('销售额最高的3个小时：\n', hourly_sales.nlargest(3))
```

# DataFrame

## 1.DataFrame的创建

```python
# 通过Series创建
s1 = pd.Series([1,2,3,4,5,6])
s2 = pd.Series([6,7,8,9,10,11,12])
df = pd.DataFrame({'第1列': s1, '第2列': s2})
print(type(df))  # DataFrame
print(type(df['第1列'])) # DataFrame的每一列都是一个Series
print(df)

# 通过字典来创建
df2 = pd.DataFrame(
    {
        "id": [1, 2, 3, 4, 5, 6],
        "name":["ada","sfds","hjgf","fdg","gdfd","fjh"],
        "age":[11,12,13,14,15,16]
    },columns=["id","age","name"] #按指定的字段排序
)
df2
```

## 2.DataFrame的属性

```python
# dataframe的属性
print("行索引：\n",df2.index)
print("列标签：\n",df2.columns)
print("值：\n",df2.values)
print("维度：\n",df2.ndim)
print("形状：\n",df2.shape)
print("数据类型：\n",df2.dtypes)
```

```python
# 获取元素 loc iloc at iat
print(df2)
## 某行
print(df2.loc[4])
## 某行(按索引)
print(df2.iloc[4])
## 某列
print(df2.loc[:,'name'])
## 某列(按索引)
print(df2.iloc[:,2])
## 多列
print(df2[['id','age']])
```

```python
## 获取单个元素
print(df2.at[2,'name'])
print(df2.iat[2,2])
```

```python
## 使用布尔索引筛选数据
#df2[(df2['id'] > 2) & (df2['age'] > 15)]
df2.id.max()
```

```python
# ======================
# 我直接给你造好测试数据
# 月份、渠道、金额
# ======================
data = {
    "月份": ["2026-01", "2026-01", "2026-01", "2026-02", "2026-02", "2026-02", "2026-03", "2026-03"],
    "渠道": ["线下门店", "线上APP", "第三方", "线下门店", "线上APP", "第三方", "线下门店", "线上APP"],
    "金额": [15000, 28000, 12000, 18000, 32000, 14000, 25000, 41000]
}

df = pd.DataFrame(data)

# 看看原始数据
print("===== 原始数据 =====")
print(df)
print("\n")

# ======================
# 你要测试的核心代码
# ======================
print("===== 透视表结果 =====")
result = df.pivot_table(index="月份", columns="渠道", values="金额", aggfunc="sum")
print(result)
```

```python
import pandas as pd

# ---------------------
# 表1：保单主表（你要补数据的表）
# ---------------------
df1 = pd.DataFrame({
    "保单号": ["P001", "P002", "P003", "P004"],
    "保费": [1000, 2000, 1500, 3000]
})

# ---------------------
# 表2：客户信息表（你要查的表）
# ---------------------
df2 = pd.DataFrame({
    "保单号": ["P001", "P002", "P003"],  # 共同字段
    "客户姓名": ["张三", "李四", "王五"],
    "渠道": ["线下", "线上", "线下"]
})

# ---------------------
# Pandas VLOOKUP 核心代码
# 用 保单号 把 客户姓名、渠道 匹配过来
# ---------------------
result = pd.merge(df1, df2, on="保单号", how="left")

print("=== 最终结果（和 VLOOKUP 一样） ===")
print(result)
```
