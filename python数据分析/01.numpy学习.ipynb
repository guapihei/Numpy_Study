# *Numpy* 学习

## 1.导包

```python
import numpy as np
```

## 2.查看属性

```python
arr = np.array([[1,2,3],[4,5,6]])
print(arr)
print('数组形状:', arr.shape)
print('元素个数:', arr.size)
print('数组维度：', arr.ndim)
print('元素数据类型:', arr.dtype)
print('转置:', arr.T)
```

## 3.*ndarry* 的创建

```python
list1 = [1,2,3],[4,5,6]
arr = np.array(list1,dtype = np.float64) #强制类型转换
print(arr)
print(arr.ndim)
```

### copy

```python
# copy
list2 = np.copy(list1)
print(list2)
arr2 = np.copy(arr)
print(arr2)
```

### 预定义形状

```python
# 预定义形状，初始化数组
# 全0
arr3 = np.zeros_like(arr2)
print(arr3)
arr4 = np.zeros((2,4))
print(arr4)
# 全1
arr5 = np.ones_like(arr2, dtype = np.int64)
print(arr5)
# 初始化为null
arr6 = np.empty_like(arr2)
print(arr6)
```

### 数列

```python
# 等差数列
arr7 = np.arange(1, 6,2) #开始，结束，步长
print(arr7)
```

```python
# 等间隔数列
arr8 = np.linspace(0,10,5,dtype = np.int64)
print(arr8)
```

### 特殊矩阵

```python
# 单位矩阵
arr = np.eye(3,4,dtype=np.int64)
print(arr)
```

```python
# 对角矩阵
list1 = [1,2,3]
arr = np.diag(list1)
print(arr)
```

```python
# 随机数组生成
## 随机浮点数（0-1）
arr = np.random.rand(2,3)
print(arr)
## 随机浮点数（指定区间）
arr2 = np.random.uniform(3,4,(2,3))
print(arr2)
## 随机整数
arr3 = np.random.randint(1,5,(2,3))
print(arr3)
## 随机数列（正态分布）
arr4 = np.random.randn(2,3)
print(arr4)
```

```python
## 设置随机种子
## 随机种子是一套固定的数学公式
## 只要种子数字一样
## 算出来的结果完全相同
np.random.seed(20)
arr5 = np.random.randint(1,5,(2,3))
print(arr5)
print(arr5)
```

## 4.*Ndarry* 的数据类型

```python
## 布尔类型
list1 = [1,0,0,1]
arr = np.array(list1,dtype = np.bool)
print(arr)

## 整数类型
list2 = [1,127,0,1]
arr = np.array(list2,dtype = np.int8) #8位有符号的整数范围为-128至127
print(arr)
list2 = [1,250,0,255]
arr = np.array(list2,dtype = np.uint8) #8位无符号的整数范围为0至255
print(arr)
```

## 5.索引与切片

```python
# 一维数组的索引与切片
np.random.seed(20)
arr = np.random.randint(1,100,10) #1-100，10个随机数
print(arr)
print(arr[1])
print(arr[1:5]) #左闭右开区间
print(arr>20) #bool类型
print(arr[(arr>20) & (arr<70)]) #bool索引，&与，|或
```

```python
## 二维数组的索引与切片
np.random.seed(20)
arr = np.random.randint(1,100,(2,3)) #1-100，10个随机数
print(arr)
print(arr[1,2])
```

## 6.*Ndarry* 的运算

```python
## 算术运算
a = np.array([1,2,3])
b = np.array([1,2,3])
print(a+b)
c = [4,5,6]
d = [7,8,9]
for i in range(0,3,1):
    print(c[i])
    print(d[i])
    d[i] = c[i] + d[i]
print(d)
```

## 7.常用函数

### 基本数学运算

```python
## 平方根
# 计算平方根
print(np.sqrt(16).astype(np.int64)) # 返回的浮点
print(np.sqrt([1,4,9])) # list每个元素求算术平方根
print(np.sqrt(np.array([1,4,9]))) # ndarray的每个元素求算术平方根
```

```python
## 指数
# 计算自然指数
print(np.exp(1)) # e的1次幂

# 计算a的b词幂
arr = np.array([1,2,3])
print(np.power(2, arr)) # 计算 2的1次幂、2次幂、3次幂
print(np.power(arr, 2)) # 计算1、2、3的2次幂
```

```python
## 对数
# 计算自然对数
print(np.log(2.71)) # 计算e为底的对数
```

```python
## 三角函数
print(np.sin(np.pi / 2))
print(np.cos(np.pi))
```

```python
## 绝对值
arr = np.array([-1, 2, -4])
print(np.abs(arr))
```

```python
## 检测缺失值
np.isnan([1,2, np.nan, 3])
```

### 统计函数

```python
## 求和、平均值、最大值、最小值等
arr = np.array([1, 2, 3])

print(np.sum([1,2,3])) # 求和
print(np.mean(arr)) # 求平均值
print(np.max(arr)) # 最大值
print(np.argmax(arr)) # 最大值所在的位置
print(np.min(arr)) # 最小值
print(np.argmin(arr)) # 最小值所在的位置
```

```python
## 获取标准差、方差
print(np.std([1,2,4,5,8,9])) # 标准差
print(np.var([1,2,4,5,8,9])) # 方差
```

### 比较函数

```python
## 数组比较大小
# 比较大于、小于、等于
arr = np.array([1,2,3,4,5,6,7])
print(arr[arr > 4])
print(arr > 4)
print(np.greater(arr, 4)) # 返回一个数组，数组值为对应元素是否大于4
print(np.less(arr, 4)) # 判断小于
print(np.equal(arr, 4)) # 判断等于

# 数组和数组比较（两个数组形状相同）
print(np.equal([3,4,5], [4,6,9])) # 返回一个数组，数组元素为两个数组的该位置元素是否相等
```

```python
## 与或非
print(np.logical_and([1,0], [1,1]))  # 逻辑与：返回一个数组。如果两个数组对应位置的元素都为True则返回True，否则返回False
print(np.logical_or([0,0], [0,1])) # 逻辑或
print(np.logical_not([1,0])) # 逻辑非：返回一个数组，将原数组对应位置元素取非
```

```python
## 检查数组元素是否为真
print(np.any([0,1,1,0])) # 返回一个布尔值：只要有一个为True，就会返回True
print(np.all([1,1,1])) # 返回一个布尔值：所有元素全部为True才返回True
```

### 排序函数

```python
## 数组排序
np.random.seed(0)
arr = np.random.randint(1, 100, 20)
print(arr)

# 直接调用ndarray数组本身的排序函数
# 该函数没有返回值，会直接修改原数组
# arr.sort()
print(arr)

# 使用 np.sort 对数组排序，会返回一个排序好的数组，不会修改原数组的内容，
print(np.sort(arr))
print(np.argsort(arr)) # 将原数字中排序的元素的索引返回
```

### 其他常用函数

```python
## 去重
np.random.seed(0)
arr = np.random.randint(1, 100, 20)

print(np.unique(arr)) # 返回一个没有重复元素的数组，并进行了排序
```

```python
## 数组拼接
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
print(np.concatenate((arr1, arr2)))  # 将两个数组拼接成一个数组
```

```python
## 数组切割
print(np.split(arr, 4))  # 将原始数组均匀的切分成4个数组。如果不能均匀切分会直接报错
print(np.split(arr, [6, 12, 18])) # 指定位置切分
```
